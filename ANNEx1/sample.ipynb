```python
class Perceptron:
    def __init__(self, input_size, learning_rate=0.1, epochs=1000):
        self.weights = np.random.randn(input_size + 1) * 0.01  # input_size + 1 for bias
        self.learning_rate = learning_rate
        self.epochs = epochs

    def activation_function(self, weighted_sum):
        if weighted_sum > 0.2:
            return 1
        elif weighted_sum < -0.2 :
            return -1
        else :
            return 0

    def predict(self, inputs):
        inputs_with_bias = np.append(inputs, 1)
        weighted_sum = np.dot(inputs_with_bias, self.weights)
        return self.activation_function(weighted_sum)

    def train(self, training_inputs, labels):
        # Training loop
        for _ in range(self.epochs):
            for inputs, label in zip(training_inputs, labels):
                prediction = self.predict(inputs)
                error = label - prediction
                inputs_with_bias = np.append(inputs, 1)  
                if error != 0:
                    self.weights += self.learning_rate * error * inputs_with_bias

if __name__ == "__main__":
    perceptron = Perceptron(input_size=63)
    d_train = data
    label = data.iloc[:, 0].values  
    #print(test_data)
    d_train_y=d_train['label']
    d_train_X=d_train.drop(['label'],axis=1)
    X=np.array(d_train_X)
    y=np.array(label)
    training_inputs = X
    print(y)
    for j in range(7):
        labels = np.array(label)
        for i in range(len(labels)):
            if labels[i] != j + 1:
                labels[i] = -1
            else :
                labels[i] = 1
        perceptron.train(training_inputs, labels)
        print(f"perceptron trained for label {j + 1} ")

        test_mask = label == 1
        test_data = data[test_mask]
        d_test_y=test_data['label']
        d_test_X=test_data.drop(['label'],axis=1)
        test_input = np.array(d_test_X)[2]
        prediction = perceptron.predict(test_input)
        print(f"Prediction for the test with label {j + 1} input: {prediction}")

    #print(f"label : {labels} ")
    # Train the perceptron
    # Test the perceptron with new data
    #print(np.array(d_test_X)[0])
    
```
